# Feature Engineering

# Categorize Age into Age groups

```python
survey_df['age_group'] = pd.cut(
    survey_df['age'],
    bins=[17, 25, 35, 45, 55, 70, float('inf')],
    labels=['18-25', '26-35', '36-45', '46-55', '56-70', '70+']
)
survey_df.drop(columns='age', inplace=True)
```

```python
survey_df
```

# Create cf_ab_score (Consume Frequency and Awareness Brand Score)

```python
frequency_mapping = {
    "0-2 times": 1,
    "3-4 times": 2,
    "5-7 times": 3
}

survey_df["frequency_score"] = survey_df["consume_frequency(weekly)"].map(frequency_mapping)
```

```python
brand_awareness_mapping = {
    "0 to 1": 1,
    "2 to 4": 2,
    "above 4": 3
}

survey_df["awareness_score"] = (
    survey_df["awareness_of_other_brands"]
    .map(brand_awareness_mapping)
)
```

```python
survey_df["cf_ab_score"] = survey_df["frequency_score"]/(survey_df["frequency_score"] + survey_df["awareness_score"])
```

```python
survey_df['cf_ab_score']=survey_df['cf_ab_score'].round(2)
```

```python
survey_df['cf_ab_score'].max()
```

# Create Zone Affluence Score (ZAS)

```python
zone_mapping = {
    "Rural": 1,
    "Semi-Urban": 2,
    "Urban": 3,
    "Metro": 4
}
survey_df["zone_score"] = survey_df["zone"].map(zone_mapping)
income_mapping = {
    "Not Reported": 0,
    "<10L": 1,
    "10L - 15L": 2,
    "16L - 25L": 3,
    "26L - 35L": 4,
    "> 35L": 5
}
survey_df["income_score"] = survey_df["income_levels"].map(income_mapping)
```

```python
survey_df['zas_score']=survey_df['zone_score']*survey_df['income_score']
```

```python
survey_df['zas_score'].nunique()
```

# Brand Switching Indicator (BSI)

```python
import numpy as np
survey_df["bsi"] = np.where(
    (survey_df["current_brand"] != "Established") &
    (survey_df["reasons_for_choosing_brands"].isin(["Price", "Quality"])),
    1,
    0
)
```

# Model Selection

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import (
    accuracy_score,
    classification_report
)
```

```python
models = {
    'Gaussian Naive Bayes': GaussianNB(),

    'Logistic Regression': LogisticRegression(
        max_iter=1000,
        random_state=42
    ),

    'Support Vector Machine': SVC(
        kernel='rbf',
        random_state=42
    ),

    'Random Forest': RandomForestClassifier(
        n_estimators=100,
        random_state=42
    ),

    'XGBoost': XGBClassifier(
        random_state=42,
        eval_metric='logloss'
    ),

    'LightGBM': LGBMClassifier(
        random_state=42,
        verbose=-1
    )
}
```

# Performance Evaluation

```python
results = []

for name, model in models.items():

    model.fit(X_train, Y_train)

    y_pred = model.predict(X_test)

    results.append({
        'Model': name,
        'Accuracy': accuracy_score(Y_test, y_pred),
        'Classification_Report': classification_report(Y_test, y_pred),
    })
```

```python
import mlflow
import mlflow.sklearn
import mlflow.xgboost
import mlflow.lightgbm
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

models = {
    "GaussianNB": GaussianNB(),
    "LogisticRegression": LogisticRegression(random_state=42, max_iter=1000),
    "SVM": SVC(probability=True, random_state=42),
    "RandomForest": RandomForestClassifier(
        n_estimators=100,
        random_state=42
    ),
    "XGBoost": XGBClassifier(
        random_state=42,
        eval_metric='logloss'
    ),
    "LightGBM": LGBMClassifier(
        random_state=42
    )
}
```

```python
mlflow.set_experiment("Beverage Price Prediction")
```

```python
for model_name, model in models.items():
    with mlflow.start_run(run_name=model_name):
        model.fit(X_train, Y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(Y_test, y_pred)
        mlflow.log_params(model.get_params())
        mlflow.log_metric("accuracy", accuracy)
        if model_name == "XGBoost":
            mlflow.xgboost.log_model(
                model,
                artifact_path="model"
            )

        elif model_name == "LightGBM":
            mlflow.lightgbm.log_model(
                model,
                artifact_path="model"
            )

        else:
            mlflow.sklearn.log_model(
                model,
                artifact_path="model"
            )

        print(f"{model_name} logged successfully.")
```
